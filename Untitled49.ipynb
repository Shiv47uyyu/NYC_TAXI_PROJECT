# maximizing revenue of taxi drivers through payment type analysis

# problem statememt
In fast-paced taxi booking sector, making the most revenue is essential for long-term success and driven happiness. Our goal is to use data-driven insights to maximize revenue streams for taxi drivers. In order to meet this need, our research aims to determine whether payment methods have an impact on fare pricing by focusing on the relationship between payment type and fare amount.






# objective
This project's main goal is to run A /B tests to examine the relationship between the total fare and the method of payment. We use Python hypothesis testing descriptive statistics to extract the useful information that can help taxi drivers generate more cash. In particular, we want to find out if there is a big difference in fares for those who pay with credit card and workers, those who pay with cash.

# reserch question
If there are relationships between total fare amount and payment type, and we can nudge customers towards payment methods that generate higher revenue for drivers, without negatively impacting customer experience?

```python
#iqr
#all of this are formula for iqr(refer notes)
#learn how to apply it no need to learn formula you will ge it on web
for col in["fare_amount","trip_distance","duration"]:
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    iqr=q3-q1
    lower_bound=q1-1.5*iqr
    upper_bound=q3+1.5*iqr
    df=df[(df[col]>=lower_bound)&(df[col]<=upper_bound)]
```
